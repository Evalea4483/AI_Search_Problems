```python
# find the shortest and longest paths between user-specified start and end nodes
# A directed graph is made up of edges going in a specific direction, which indicates
# a one- way relationship beween two nodes

# import libraries
import networkx as nx
import matplotlib.pyplot as plt

# initialize a directed graph using the DiGraph class from the networkx library
G = nx.DiGraph()

#set up a graph with 5 nodes, each node as a specific label 'A' through 'F'
nodes = ['A', 'B', 'C', 'D', 'E', 'F']
G.add_nodes_from(nodes)

# define directed edges
edges = [
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'D'),
    ('B', 'E'),
    ('C', 'F'),
    ('D', 'A'),
    ('F', 'B'),
]

# add directed edgs between nodes
G.add_edges_from(edges)

# visualize the directed graph
pos = nx.spring_layout(G)
plt.figure(figsize=(8,8))
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=10, font_color='black', font_weight = 'bold')
plt.title("Directed Graph")
plt.show()

num_nodes = G.number_of_nodes()
print("Number of nodes in the graph", num_nodes)

# get user input for the start and goal node (state)
start_node = input("Enter the start node: ")
goal_node = input("Enter the goal node: ")

# path finding function
def find_paths(graph, start, goal):
    try:
        shortest_paths = list(nx.all_shortest_paths(graph, source=start, target=goal))
        longest_path = max(list(nx.all_simple_paths(graph, source=start, target=goal)), key= len)
        return shortest_paths, longest_path
    except nx.NetworkXNoPath:
        return None, None

# unpack the results of the path finding function into the shortest and longest paths
shortest_paths, longest_path = find_paths(G, start_node, goal_node)

# if the shortest path exists, print the shortest path between those nodes
if shortest_paths:
    print(f"Your selected path from {start_node} to {goal_node}: {shortest_paths[0]}")
    print (f"Shortest paths from {start_node} to {goal_node}:")
    for path in shortest_paths:
        print(path)
else:
    print(f"No shortest path found from {start_node} to {goal_node}")
    
# if the longest path exists, print the longest path between those nodes
if longest_path:
    print(f"longest path from {start_node} to {goal_node}: {longest_path}")
else:
    print(f"No longest path found from {start_node} to {goal_node}")
```

```python
# search problem using directed graph consisting of 5 nodes where the least_cost path is one of many arcs

import networkx as nx
import matplotlib.pyplot as plt
# define arcs

arcs = [
    ('A', 'B', 3), ('A', 'C', 1),
    ('B', 'D', 1), ('B', 'G', 3),
    ('C', 'B', 1), ('C', 'D', 3),
    ('D', 'G', 1)
]

# create directed graph
search_graph = nx.DiGraph()

# add nodes and edges
for start, end, weight in arcs:
    search_graph.add_edge(start, end, weight=weight)
    
# define node positions
positions = {'A': (0,1), 'B': (0.5, 0.5), 'C': (0, 0.5), 'D': (0.5, 0), 'G': (1,0)}

# draw graph with labels
nx.draw(search_graph, pos=positions, with_labels=True, node_size=2000, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", arrowsize=20)

# edge labels
edge_labels = {(start, end): weight['weight'] for start, end, weight in search_graph.edges(data=True)}
nx.draw_networkx_edge_labels(search_graph, pos=positions, edge_labels=edge_labels)

num_nodes = search_graph.number_of_nodes()
print("Number of nodes in the graph:", num_nodes)

# visualize the graph
plt.show()
```
